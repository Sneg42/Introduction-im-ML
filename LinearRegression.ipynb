```python
import mglearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
%matplotlib inline
from IPython.display import display
plt.rc('font', family='Verdena')
```

```python
mglearn.plots.plot_linear_regression_wave()
```

```python
from sklearn.linear_model import LinearRegression
X, y = mglearn.datasets.make_wave(n_samples=60)
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42)

lr = LinearRegression().fit(X_train, y_train)
```

```python
print ('lr.coef_: {}'.format(lr.coef_))
print ('lr.intercept_: {}'.format(lr.intercept_))
print ('{:.2f}'.format(lr.score(X_train, y_train)))
print ('{:.2f}'.format(lr.score(X_test, y_test)))
```

```python
X, y = mglearn.datasets.load_extended_boston()
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=0)

lr = LinearRegression().fit(X_train, y_train)
```

```python
from sklearn.linear_model import Ridge

ridge = Ridge().fit(X_train, y_train)
print('{:.2f}'.format(ridge.score(X_train, y_train)))
print('{:.2f}'.format(ridge.score(X_test, y_test)))
```

```python
ridge10 = Ridge(alpha=10).fit(X_train, y_train)
print('{:.2f}'.format(ridge10.score(X_train, y_train)))
print('{:.2f}'.format(ridge10.score(X_test, y_test)))
```

```python
ridge01 = Ridge(alpha=0.1).fit(X_train, y_train)
print('{:.2f}'.format(ridge01.score(X_train, y_train)))
print('{:.2f}'.format(ridge01.score(X_test, y_test)))
```

```python
plt.plot(ridge.coef_, 's', label="Гребневвая регрессия alpha=1")
plt.plot(ridge10.coef_, 's', label="Гребневвая регрессия alpha=10")
plt.plot(ridge01.coef_, 's', label="Гребневвая регрессия alpha=0.1")

plt.plot(lr.coef_, 'o', label='Линейная регрессия')
plt.xlabel('Индекс коэффициента')
plt.ylabel('Оценка коэффициента')
plt.hlines(0, 0, len(lr.coef_))
plt.ylim(-25, 25)
plt.legend()
```

```python
mglearn.plots.plot_ridge_n_samples()
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso().fit(X_train, y_train)
print('Правильность на обучающей выборке: {:.2f}'.format(lasso.score(X_train, y_train)))
print('Правильность на контрольной выборке: {:.2f}'.format(lasso.score(X_test, y_test)))
print('Количесвто использованых признаков: {}'.format(np.sum(lasso.coef_ != 0)))
```

```python
lasso001 = Lasso(alpha=0.01, max_iter=100000).fit(X_train, y_train)
lasso00001 = Lasso(alpha=0.0001, max_iter=100000).fit(X_train, y_train)
print('Правильность на обучающей выборке: {:.2f}'.format(lasso001.score(X_train, y_train)))
print('Правильность на контрольной выборке: {:.2f}'.format(lasso001.score(X_test, y_test)))
print('Количесвто использованых признаков: {}'.format(np.sum(lasso001.coef_ != 0)))
```

```python
plt.plot(lasso.coef_, 's', label="Лассо alpha=1")
plt.plot(lasso001.coef_, '^', label="Лассо alpha=0.01")
plt.plot(lasso00001.coef_, 'v', label="Лассо alpha=0.0001")

plt.plot(ridge.coef_, 'o', label="Гребневая регрессия alpha=0.0001")
plt.legend(ncol=2, loc=(0, 1.05))
plt.ylim(-25, 25)
plt.xlabel("Индекс коэффициента")
plt.ylabel("Оценка коэффициента")
```
